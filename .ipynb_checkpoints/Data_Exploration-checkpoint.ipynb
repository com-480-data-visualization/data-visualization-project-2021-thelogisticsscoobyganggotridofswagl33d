**Data Exploration of the Spotify Dataset**


```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
french_tracks = pd.read_csv("FR_tracks.csv")
french_artists = pd.read_csv("FR_artists.csv")
french_tracks.head()
```

```python
print("Tre")
french_tracks[french_tracks.release_date.apply(lambda x : len(x)) > 4] #.groupby("release_date").count().sort_values("id", ascending=False).head(50)
```

```python
french_tracks.release_date = french_tracks.release_date.apply(lambda x : x[:4])
french_tracks
```

```python
b = french_tracks.groupby('release_date').mean()
c = french_tracks.groupby('release_date').count()


plt.figure(figsize=(15,6))
plt.plot(c, label= "count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.legend()
plt.show()

c
```

```python
plt.figure(figsize=(15,6))
plt.plot(b.energy, label= "energy")
plt.xticks(rotation=90)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.plot(b.danceability, label= "danceability")
plt.xticks(rotation=90)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.plot(b.popularity, label= "popularity")
plt.xticks(rotation=90)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
french_artists
```

```python
genres = french_artists.sort_values("popularity", ascending=False).genres.apply(lambda x : np.asarray(x.replace("[","").replace("]","").replace("'","").strip().split(","))).values
nb_empty = 0 
genrs = {}
for g in genres :
    if g.size == 1 :
        if g == '':
            nb_empty+=1
    for s in g:
        if s not in genrs:
            genrs[s] = 1
        else :
            genrs[s] += 1
            
print(nb_empty)
genrs
```

```python
a = french_artists.groupby('year').mean()

plt.plot(a.popularity)
plt.show()
```

***Billboard Dataset***

```python
billboard_df = pd.read_csv("BillboardFromLast20/billboardHot100_1999-2019.csv")
```

```python
billboard_cleaned = billboard_df.drop("Features",1).dropna()
billboard_cleaned
```

```python

genres = {}
for g in billboard_cleaned.Genre :
    for g_ in g.split(","):
        if g_ not in genres : 
            genres[g_] = 1
        else :
            genres[g_] += 1
            
            
ye = pd.DataFrame.from_dict(genres, orient='index').sort_values(0, ascending=False)
ye.columns = ['nb_songs']
#sns.scatterplot(x= range(307), y = range(307), size = ye.nb_songs)
ye
```
